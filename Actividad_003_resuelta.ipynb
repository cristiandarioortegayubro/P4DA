> **Y por Ãºltimo, hacer un grÃ¡fico financiero de velas... para las acciones de Google...** ðŸ˜ˆðŸ˜ˆðŸ˜ˆ
</font>

````python
GOOG = yf.download("GOOG", start="2022-01-01", end="2022-03-31").round(2)
GOOG.reset_index(inplace=True)
GOOG["Date"] = pd.to_datetime(GOOG.Date.dt.date)
GOOG
````

````python
import plotly.graph_objects as go
````

````python
fig = go.Figure(data=[go.Candlestick(x=GOOG["Date"],
                open=GOOG['Open'],
                high=GOOG['High'],
                low=GOOG['Low'],
                close=GOOG['Close'])])

fig.update_layout(template="gridon",
                  title="CotizaciÃ³n de Google")


fig.show()
````
