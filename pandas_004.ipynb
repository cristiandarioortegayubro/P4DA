## **<font color="RoyalBlue">Para análisis de datos</font>**

```python
import pandas as pd
import numpy as np
```

## **<font color="RoyalBlue">Para acceso remoto de datos</font>**


```python
import yfinance as yf 
```

## **<font color="RoyalBlue">Para dar aspecto plotly a los gráficos**

```python
pd.options.plotting.backend = "plotly"
```

# **<font color="RoyalBlue">Obtención de datos**

```python
datos = "https://raw.githubusercontent.com/cristiandarioortegayubro/BA/main/Datasets/Analisis_de_Facturas_de_Venta.csv"
df = pd.read_csv(datos).round(2)
df["FECHA"] = pd.to_datetime(df.FECHA)
df.drop(columns=["DOCUMENTO","TIPOFACTURA"], inplace=True)
df = df[df.MONEDA == "Pesos Argentinos"]
df
```

```python
df = df.sort_values(by="FECHA", ascending=True)
df = df.reset_index(drop=True)
df
```

# **<font color="RoyalBlue">Usando Plot de Pandas**

```python
clientes = df.groupby("ORGANIZACION").sum()
clientes.head()
```

```python
clientes.reset_index(inplace=True)
clientes.head()
```

```python
clientes.plot("ORGANIZACION",
              "IMPORTEMONTRANSACCION", 
              kind="bar",
              color="ORGANIZACION",
              title="Ventas totales por cliente",
              labels={"IMPORTEMONTRANSACCION":"Ventas","ORGANIZACION":"Clientes"})
```

# **<font color="RoyalBlue">Obtención de datos financieros**

```python
dolar = yf.download("ARS=X", start="2022-01-03", end="2022-01-21").round(2)
dolar.drop(columns=["Volume","Adj Close"], inplace=True)
dolar.reset_index(inplace=True)
dolar.head(3)
```

```python
bitcoin = yf.download("BTC-USD", start="2022-01-01", end="2022-01-14").round(2)
bitcoin.drop(columns=["Volume","Adj Close"], inplace=True)
bitcoin.reset_index(inplace=True)
bitcoin.head()
```

# **<font color="RoyalBlue">Usando Plot de Pandas**

## **<font color="RoyalBlue">Gráfico Lineal**

```python
dolar.plot("Date","Close", 
           kind="line", 
           title="COTIZACIÓN DOLAR",
           labels={"Close":"Cotizacion de Cierre","Date":"Fecha"},
           markers=True)
```

```python
bitcoin.plot("Date","Close", 
            kind="line",
            template="simple_white", 
            title="COTIZACIÓN BITCOIN")
```

```python
bitcoin.plot("Date",["Open","Close"], 
            kind="line",
            title="COTIZACIÓN BITCOIN")
```

```python
bitcoin = yf.download("BTC-USD", start="2022-01-01", end="2022-01-14").round(2)
bitcoin.drop(columns=["Volume","Adj Close"], inplace=True)
bitcoin.head()
```

```python
bitcoin.plot(facet_row="variable",
             template="gridon",
             title="COTIZACION BITCOIN")
```

```python
bitcoin.plot.line()
```

```python
bitcoin.plot.line(template="gridon", title="COTIZACION BITCOIN")
```

## **<font color="RoyalBlue">Gráfico de Caja**

```python
bitcoin.plot.box()
```

```python
bitcoin.plot.box(template="gridon", title="BOXPLOT BITCOIN")
```

```python
bitcoin.plot.box(template="gridon",
                 color="variable", 
                 title="BOXPLOT BITCOIN")
```
