# <font color="RoyalBlue">**Habilitando bibliotecas requeridas**</font>

## **<font color="RoyalBlue">Para análisis de datos</font>**

```python
import pandas as pd
import numpy as np
```

## **<font color="RoyalBlue">Para gráficos**

```python
import plotly.express as px
```

# **<font color="RoyalBlue">Obtención de datos**

```python
url = "https://raw.githubusercontent.com/LucaAPiattelli/Diplomatura_Business_Analytics_UDA/main/Modulo_08_Aprendizaje_Automatico/visualizacion.csv"
```

# **<font color="RoyalBlue">Creación del dataframe**

```python
analisis = pd.read_csv(url)
analisis.head()
```

```python
analisis.columns
```

# **<font color="RoyalBlue">Preparación de los datos**

## **<font color="RoyalBlue">Renombrando las columnas**

**Se cambian los nombres de las columnas para mejor interpretacion...**

```python
analisis.info()
```

```python
analisis.rename(columns={"satisfaction_level":"niveldesatisfaccion",
                         "last_evaluation":"ultimaevaluacion",
                         "number_project":"numerosdeproyectos",
                         "average_montly_hours":"horasmensualespromedio",
                         "time_spend_company":"tiempoenlaempresa",
                         "Work_accident":"accidentedetrabajo",
                         "left":"abandono",
                         "promotion_last_5years":"promocionultimos5años",
                         "sales":"ventas",
                         "salary":"sueldo"}, inplace= True)
```

```python
analisis.info()
```

## **<font color="RoyalBlue">Visualizando los datos**

```python
px.scatter(analisis, 
           x="niveldesatisfaccion", 
           y="ultimaevaluacion",
           color="abandono",
           template="gridon",
           title="Nivel de Satisfacción vs. Ultima Evaluacion",
           labels={"niveldesatisfaccion":'Nivel de Satisfacción',
                   "ultimaevaluacion":"Ultima Evaluación",
                   "abandono":"Abandono"})
```

```python
px.scatter(analisis, 
           x="niveldesatisfaccion", 
           y="ultimaevaluacion",
           color_continuous_scale="mint",
           template="gridon",
           color="abandono",                   
           labels={"niveldesatisfaccion":'Nivel de Satisfacción',
                   "ultimaevaluacion":"Ultima Evaluación",
                   "abandono":"Abandono"})
```

```python
px.histogram(analisis, 
             barmode="relative",
             title="Nivel de Satisfacción",
             template="gridon",
             x="niveldesatisfaccion",
             labels={"niveldesatisfaccion":"Nivel de Satisfacción"}).update_layout(bargap=0.2)
```

```python
px.histogram(analisis, 
                   barmode="relative",
                   title="Ultima Evaluación",
                   template="gridon",
                   x="ultimaevaluacion",
                   labels={"ultimaevaluacion":"Ultima Evaluación"}).update_layout(bargap=0.2)
```

```python
px.histogram(analisis, x="sueldo",
             title="Sueldos",
             color="sueldo",
             template="gridon")
```

```python
px.histogram(analisis, x="sueldo",
             title="Sueldos",
             color="abandono",
             template="gridon")
```

```python
px.scatter(analisis, 
           x="niveldesatisfaccion", 
           y="ultimaevaluacion",
           color="abandono",
           template="gridon",
           marginal_x="histogram",
           marginal_y="histogram",
           labels={"niveldesatisfaccion":'Nivel de Satisfacción',
                   "ultimaevaluacion":"Ultima Evaluación"})
```

```python
px.scatter(analisis, 
           x="niveldesatisfaccion", 
           y="ultimaevaluacion",
           color="abandono",
           template="gridon",
           marginal_x="violin",
           marginal_y="violin",
           labels={"niveldesatisfaccion":'Nivel de Satisfacción',
                   "ultimaevaluacion":"Ultima Evaluación"})
```

## **<font color="RoyalBlue">Análisis de correlación**

```python
analisiscorrelacion = analisis.corr().round(2)
```

```python
analisiscorrelacion
```

```python
px.imshow(analisiscorrelacion,
          title = "Matriz de correlacion",
          text_auto = True, 
          template = "gridon",
          labels={"color":"Coeficiente"})
```
