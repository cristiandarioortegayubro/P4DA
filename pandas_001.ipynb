## 游댍 **<font color="RoyalBlue">Selecci칩n por etiquetas</font>**

<p align=justify><b><font color="RoyalBlue">Pandas proporciona un conjunto de m칠todos para tener una indexaci칩n basada en etiquetas. Este es un protocolo estricto basado en la inclusi칩n. Cada etiqueta solicitada debe existir, tanto en el index como en las columnas. Al seleccionar por ejemplo filas (index) se incluye el l칤mite inicial indicado y el l칤mite final. Los n칰meros enteros son etiquetas v치lidas, pero se refieren a las etiquetas del index (cuando la etiqueta es num칠rica) y no a la posici칩n de la fila.</font></b></p>

<p align=justify><b><font color="RoyalBlue">En los par치metros de la selecci칩n, siempre se indica primero las filas y luego las columnas.. En el caso de seleccionar de un DataFrame, se hace de la siguiente forma:</font></b></p>

~~~python
df.loc[fila inicial : fila final, columna inicial : columna final]
~~~


<p align=justify><b><font color="RoyalBlue">

> **Seleccionando tres filas, desde la fila 0 hasta la fila 2...**
</font></b></p>

```python
df.loc[0:2]
```

<p align=justify><b><font color="RoyalBlue">

> **Seleccionando cuatro filas, desde la fila 2 hasta la fila 5...**

```python
df.loc[2:5]
```

```python
df.loc[2:5,]
```

```python
df.loc[2:5,:]
```

<p align=justify><b><font color="RoyalBlue">

## 游댍 **<font color="RoyalBlue">Selecci칩n por posici칩n</font>**

<p align=justify><b><font color="RoyalBlue">Pandas proporciona un conjunto de m칠todos para obtener una indexaci칩n puramente basada en enteros. La sem치ntica sigue de cerca lo definido en Python y NumPy. Estos son indexaci칩n. Al cortar, se incluye el l칤mite inicial $n$, mientras que se excluye el l칤mite final $m$, es decir desde $n$ hasta $m-1$</font></b></p>

<p align=justify><b><font color="RoyalBlue">En los par치metros de la selecci칩n, siempre se indica primero las filas y luego las columnas. En el caso de seleccionar de un DataFrame, se hace de la siguiente forma:</font></b></p>

~~~python
df.loc[fila posici칩n inicial : fila posici칩n final, columna posici칩n inicial : columna posici칩n final]
~~~

<p align=justify><b><font color="RoyalBlue">

> **Selecci칩n por ubicaci칩n del dato**
</font></b></p>

```python
df.iloc[6]
```

```python
df.iloc[3:5,0:2]
```

<p align=justify><b><font color="RoyalBlue">

> **Tambi칠n se pueden presentar las posiciones de las filas y de las columnas como una lista...**
</font></b></p>

```python
df.iloc[[0,2,4],[0,2,3]]
```

```python
df.iloc[[1,2,4],[0,2]]
```

## 游댍 **<font color="RoyalBlue">Encadenamiento</font>**

## 游댍 **<font color="RoyalBlue">Filtros</font>**

<p align=justify><b><font color="RoyalBlue">Podemos llevar a cabo algunas opciones para filtrar nuestros datos, incluidas las m치scaras booleanas y algunos m칠todos especiales. Con las m치scaras booleanas, probamos nuestros datos contra alg칰n valor y obtenemos una estructura de la misma forma, excepto que est치 llena de valores que son True o False.</font></b></p>

```python
df.head(3)
```

<p align=justify><b><font color="RoyalBlue">

> **Filtrando los datos para valores de PRECIO mayores a $100000$, en este caso, se muestra la serie PRECIO con datos booleanos que cumplen la condici칩n...**
</font></b></p>

```python
df.PRECIO > 100000
```

<p align=justify><b><font color="RoyalBlue">

> **Si lo que se quiere es ver el resultante de la condici칩n en todo el DataFrame, entonces...**
</font></b></p>

```python
df[df.PRECIO > 100000]
```

<p align=justify><b><font color="RoyalBlue">

>**Para visualizar m치s de dos condiciones verdaderas, es decir todas las condiciones verdaderas se usa el s칤mbolo &...**
</font></b></p>

```python
df.loc[(df.CANTIDAD == 1) & (df.PRECIO > 110000)]
```

<p align=justify><b><font color="RoyalBlue">

>**Para visualizar m치s de dos condiciones y alguna de ellas es verdadera, se usa el s칤mbolo |**
</font></b></p>

```python
df.loc[(df.CANTIDAD > 50) | (df.PRECIO > 110000)]
```

<p align=justify><b><font color="RoyalBlue">

>**Para un conjunto de valores entre un intervalo determinado...**
</font></b></p>

```python
df.loc[df.PRECIO.between(100000, 150000)]
```

```python
df.loc[df.PRECIO.between(10000, 110000) & df.CANTIDAD.between(1, 30)]
```

```python
df.loc[df.PRECIO.between(10000, 50000) | df.CANTIDAD.between(1, 4)]
```
