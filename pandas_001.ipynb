# **<font color="RoyalBlue">Carga de módulos necesarios**</font>

```python
import pandas as pd
import numpy as np
```

# **<font color="RoyalBlue">Obtención de datos</font>**

```python
datos = "https://raw.githubusercontent.com/cristiandarioortegayubro/BA/main/Datasets/Analisis_de_Facturas_de_Venta.csv"
```

<p align=justify><b><font color="RoyalBlue">Usamos el siguiente método de pandas:</font></b></p>

~~~python
pd.read_csv()
~~~


```python
df = pd.read_csv(datos)
df.info()
```

👀 **Podemos observar que, como el tipo de datos de la fecha no es el que corresponde...**

# **<font color="RoyalBlue">Selección de datos...</font>**

## 🔎 **<font color="RoyalBlue">Selección por etiquetas</font>**

<p align=justify><b><font color="RoyalBlue">Pandas proporciona un conjunto de métodos para tener una indexación basada en etiquetas. Este es un protocolo estricto basado en la inclusión. Cada etiqueta solicitada debe existir, tanto en el index como en las columnas. Al seleccionar por ejemplo filas (index) se incluye el límite inicial indicado y el límite final. Los números enteros son etiquetas válidas, pero se refieren a las etiquetas del index (cuando la etiqueta es numérica) y no a la posición de la fila.</font></b></p>

<p align=justify><b><font color="RoyalBlue">En los parámetros de la selección, siempre se indica primero las filas y luego las columnas.. En el caso de seleccionar de un DataFrame, se hace de la siguiente forma:</font></b></p>

~~~python
df.loc[fila inicial : fila final, columna inicial : columna final]
~~~


<p align=justify><b><font color="RoyalBlue">

> **Seleccionando tres filas, desde la fila 0 hasta la fila 2...**
</font></b></p>

```python
df.loc[0:2]
```

<p align=justify><b><font color="RoyalBlue">

> **Seleccionando cuatro filas, desde la fila 2 hasta la fila 5...**
</font></b></p>

```python
df.loc[2:5]
```

```python
df.loc[2:5,]
```

```python
df.loc[2:5,:]
```

```python
df.loc[2:5,"DOCUMENTO":"CANTIDAD"]
```

<p align=justify><b><font color="RoyalBlue">

> **Seleccionando dos columnas, TIPOFACTURA y FECHA...**
</font></b></p>

```python
df.loc[:,["TIPOFACTURA","FECHA"]]
```

<p align=justify><b><font color="RoyalBlue">

> **Seleccionando filas y columnas**
</font></b></p>

```python
df.loc[7:9,["FECHA","TIPOFACTURA"]]
```

```python
df.loc[[7,9],["FECHA","TIPOFACTURA"]]
```

```python
df.loc[[7,9,8],["FECHA","TIPOFACTURA"]]
```

```python
df.loc[:,["FECHA","TIPOFACTURA"]].head()
```

## 🔎 **<font color="RoyalBlue">Selección por posición</font>**

<p align=justify><b><font color="RoyalBlue">Pandas proporciona un conjunto de métodos para obtener una indexación puramente basada en enteros. La semántica sigue de cerca lo definido en Python y NumPy. Estos son indexación. Al cortar, se incluye el límite inicial $n$, mientras que se excluye el límite final $m$, es decir desde $n$ hasta $m-1$</font></b></p>

<p align=justify><b><font color="RoyalBlue">En los parámetros de la selección, siempre se indica primero las filas y luego las columnas. En el caso de seleccionar de un DataFrame, se hace de la siguiente forma:</font></b></p>

~~~python
df.loc[fila posición inicial : fila posición final, columna posición inicial : columna posición final]
~~~

<p align=justify><b><font color="RoyalBlue">

> **Selección por ubicación del dato**
</font></b></p>

```python
df.iloc[6]
```

```python
df.iloc[3:5,0:2]
```

<p align=justify><b><font color="RoyalBlue">

> **También se pueden presentar las posiciones de las filas y de las columnas como una lista...**
</font></b></p>

```python
df.iloc[[0,2,4],[0,2,3]]
```

```python
df.iloc[[1,2,4],[0,2]]
```

## 🔎 **<font color="RoyalBlue">Encadenamiento</font>**

<p align=justify><b><font color="RoyalBlue">Con esta forma de seleccionar datos, primero se deben elegir los nombres de las columnas deseadas y luego el rango de filas a mostrar...</font></b></p>

```python
df[["DOCUMENTO", "FECHA"]][0:4]
```

```python
df[["FECHA", "PRODUCTO", "PRECIO", "DOCUMENTO"]][0:10]
```

## 🔎 **<font color="RoyalBlue">Filtros</font>**

<p align=justify><b><font color="RoyalBlue">Podemos llevar a cabo algunas opciones para filtrar nuestros datos, incluidas las máscaras booleanas y algunos métodos especiales. Con las máscaras booleanas, probamos nuestros datos contra algún valor y obtenemos una estructura de la misma forma, excepto que está llena de valores que son True o False.</font></b></p>

```python
df.head(3)
```

<p align=justify><b><font color="RoyalBlue">

> **Filtrando los datos para valores de PRECIO mayores a $100000$, en este caso, se muestra la serie PRECIO con datos booleanos que cumplen la condición...**
</font></b></p>

```python
df.PRECIO > 100000
```

<p align=justify><b><font color="RoyalBlue">

> **Si lo que se quiere es ver el resultante de la condición en todo el DataFrame, entonces...**
</font></b></p>

```python
df[df.PRECIO > 100000]
```

<p align=justify><b><font color="RoyalBlue">

>**Para visualizar más de dos condiciones verdaderas, es decir todas las condiciones verdaderas se usa el símbolo &...**
</font></b></p>

```python
df.loc[(df.CANTIDAD == 1) & (df.PRECIO > 110000)]
```

<p align=justify><b><font color="RoyalBlue">

>**Para visualizar más de dos condiciones y alguna de ellas es verdadera, se usa el símbolo |**
</font></b></p>

```python
df.loc[(df.CANTIDAD > 50) | (df.PRECIO > 110000)]
```

<p align=justify><b><font color="RoyalBlue">

>**Para un conjunto de valores entre un intervalo determinado...**
</font></b></p>

```python
df.loc[df.PRECIO.between(100000, 150000)]
```

```python
df.loc[df.PRECIO.between(10000, 110000) & df.CANTIDAD.between(1, 30)]
```

```python
df.loc[df.PRECIO.between(10000, 50000) | df.CANTIDAD.between(1, 4)]
```
